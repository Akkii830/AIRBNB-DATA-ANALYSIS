## **Question 5) What are the total number of Hosts that provides Minimum night stays?**

```python
#First we make the minimum night category.
def minimum_night_category(x):
  if x==1:
    return 'one day'
  if x>1 and x<=3:
    return '2-3 days'
  if x>3 and x<7:
    return 'few days'
  if x==7:
    return '1 week'
  elif x>7 and x<=29:
    return 'less than a month'
  elif x==30:
    return '1 month'
  elif x>30 and x<180:
    return 'more than a month'
  elif x>=180 and x<=365:
    return 'more than half year'
  elif x>365:
    return 'more than a year'  

#Creating a new column for the minimum night categories.
AirBnb_df['minimum_night_category'] = AirBnb_df['minimum_nights'].apply(minimum_night_category)
```

```python
#Now make a Dataframe of the minimum night category and its total number of hosts.
mnc_tnh = pd.DataFrame(AirBnb_df.groupby('minimum_night_category')['id'].count()).reset_index()
mnc_tnh.sort_values('id', ascending = False)
mnc_tnh
```

```python
#Ploting graph
w=0.5
plt.figure(figsize=(9,6))
bar1=np.arange(9)
plt.bar(bar1,mnc_tnh['count'],color='b')
plt.xlabel('type of minimum night category',size=14)
plt.ylabel('number of host',size=14)
plt.xticks(bar1,mnc_tnh['minimum_night_category'],rotation=90,fontsize=11)
plt.legend()
plt.show()
```
